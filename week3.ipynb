```python
#simulate a quiz program
print("when did india get its independence?")
year = int(input())

while(year!=1947):
  print("you got this wrong, enter once again.")
  year = int(input())

print("hip hip hurray. you got it right!")
```

```python
#factorial of a number

print("enter a number :")
n = int(input())
i=1
ans=1

while(i<=n):
  ans = ans*i
  i=i+1

print(ans)
```

```python
print("enter a number :")
n = int(input())

fact=1
while(n>0):
  fact = fact*n
  n = n-1

print(fact)
```
