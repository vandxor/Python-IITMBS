```python
#typecasting : conversion of one data type into another

a = int(9.87)   #float to int
b = int('89')   #string to int

print(a, type(a))
print(b, type(b))
```

```python
c = float(80)   #int to float
d = float('67') #string to float

print(c, type(c))

print(d, type(d))
```

```python
e = str(87)  #int to string
f = str(8.65) #float to string

print(e, type(e))
print(f, type(f))
```

```python
 #boolean ka typecasting

 a = bool(10)
 b = bool(0)
 c = bool(-10)

 print(a, type(a))
print(b, type(b))
print(c, type(c))

#only zero is false


#string to boolean

d = bool("India")
print(d, type(d))

e = bool('0')
f = bool(0.0)
print(e, type(e))
print(f, type(f))

g = bool("") #empty string
print(g, type(g))
```
