````python
#lower()

#converts every character of string into lower case

x="VANDougH cookieS"
print(x.lower())
````

````python
#upper()

#converts every character of string into upper case

y="Lmaoooo"
print(y.upper())
````

````python
#capitalize
#converts first character of string to upper case
#except that letter, all remaining charactters get converted to lowercase

print(x.capitalize())
````

````python
#title()

#converts first character of everyword in that string to uppercase
#except fist letter of em words, all other letters get converted to lowercase characters

print(x.title())
````

````python
#swapcase()

#swaps cases, swaps upper case to lower case and vice versa

print(x.swapcase())
````

````python
#islower()

#returns true if all characters are lowercase
#we are not converting, we are only checking
p="python"

print(x.islower())
print(p.islower())
````

````python
#isupper()

#returns true if all characters are uppercase
#we are not converting, we are only checking
p="PYTHON"

print(x.isupper())
print(p.isupper())
````

**STRIP METHODS**

they return trimmed versions of string



````python
'''

removes spaces at beginning and end of string, it removes leading or trailing chars


strip() - returns a trimmed ver of string
lstrip() - returns a left trim version of string
rstrip() - returns a right trim version of string


'''


c = "----python----"
print(c.strip("-p"))
print(c.lstrip("-p"))
print(c.rstrip("-p"))

#u cant remove p without removing -
#order mein nikalna padta, order of trailing or lweading
print("\n")

b="   ohio"
print(b)
print(b.strip())
print(b.lstrip())
print(b.rstrip())
````

````python
'''






startswith() - returns true if string starts w specified value

endswith() - returns true if string ends w specified value

ts case sensitive

'''

o = "plone"

print(o.startswith("pl"))
print(o.endswith("ne"))
print(o.startswith("P"))
print(o.endswith("E"))
````

````python
'''

count() - returns no. of times specified value occurs in string

index() - searches for value (left to right) and returns its index

replace() - replaces one value w another, same as find and replace feature



ts case sensitive


'''
#x="VANDougH cookieS"

print(x.count('o'))
print(x.count('d'))
print("\n")

print(x.index("V"))
print(x.index("c"))
print(x.index(" "))
#print(x.index("b"))  - gives eroor
print("\n")


t = "tart"

print(t.replace("t", "f"))
print(t.replace("ta", "fa"))
print(t.replace("x", "f"))
print(t.replace("T", "f"))
print("\n")
````
